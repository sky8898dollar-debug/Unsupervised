 Q 1 What is Dimensionality Reduction? Why is it important in machine
learning?


Ans-Dimensionality reduction is the process of reducing the number of input variables or features in a dataset. It is crucial in machine learning for several key reasons. First, it tackles the "Curse of Dimensionality," where high-dimensional data becomes sparse, making models less effective. Second, it significantly reduces computational cost and training time. Third, it helps in removing noise and redundant features, leading to more robust models. Finally, it allows for the visualization of complex, high-dimensional data in 2D or 3D plots, enabling better understanding and insight.

Q 2 Name and briefly describe three common dimensionality reduction
techniques.

Ans- Of course. Here are three common dimensionality reduction techniques described briefly.

**1. Principal Component Analysis (PCA)**
PCA is a linear technique that identifies the axes (principal components) in the data that capture the maximum variance. It projects the data onto a lower-dimensional subspace while preserving as much of the data's variation as possible, effectively creating new, uncorrelated features.

**2. t-Distributed Stochastic Neighbor Embedding (t-SNE)**
t-SNE is a non-linear technique ideal for visualization. It works by converting high-dimensional Euclidean distances between data points into conditional probabilities representing similarities. It then creates a low-dimensional map where similar points are modeled by nearby points, effectively revealing clusters.

**3. Uniform Manifold Approximation and Projection (UMAP)**
UMAP is a modern non-linear technique that assumes the data is uniformly distributed on a Riemannian manifold. It works by building a topological representation of the high-dimensional data and then finding a low-dimensional projection that best preserves this structure, often outperforming t-SNE in speed and preserving more global structure.

Q 3  What is clustering in unsupervised learning? Mention three popular
clustering algorithms.

Ans-

Clustering is an unsupervised learning technique used to group unlabeled data points based on their similarities. The goal is to segment data so that items in the same group (cluster) are more alike to each other than to those in other groups. It discovers inherent patterns and structures without any prior training.

Three popular clustering algorithms are:
1.  **K-Means:** Partitions data into 'K' distinct, non-overlapping clusters based on distance from central points (centroids).
2.  **Hierarchical Clustering:** Builds a tree-like hierarchy of clusters, often shown as a dendrogram, allowing analysis at different levels of granularity.
3.  **DBSCAN:** Groups together densely packed points, effectively identifying clusters of arbitrary shapes and marking outliers in sparse regions.

Q 4 : Explain the concept of anomaly detection and its significance.

Ans-

Anomaly detection is the process of identifying rare items, events, or patterns that significantly deviate from the majority of data. These deviations are called outliers or anomalies. Its core significance lies in finding critical, unusual information that often indicates a problem or a valuable opportunity.

For instance, it is crucial for:
*   **Security:** Detecting fraudulent credit card transactions.
*   **Safety:** Identifying a faulty component in manufacturing.
*   **Health:** Spotting unusual network traffic that signals a cyberattack.

By automatically flagging these exceptions, it enables proactive measures, prevents losses, and ensures system integrity.

Q 5 : List and briefly describe three types of anomaly detection techniques.

Ans- Of course. Here are three common types of anomaly detection techniques:

1.  **Statistical Methods:** These techniques assume a distribution for the data (e.g., Gaussian) and identify anomalies as data points that fall in the low-probability regions of the model. For instance, any point beyond three standard deviations from the mean might be flagged.

2.  **Clustering-Based Methods:** This approach groups similar data points into clusters. Anomalies are points that do not belong strongly to any cluster, either lying far away from all cluster centers or residing in very small, sparse clusters, indicating they are outliers.

3.  **Reconstruction-Based Methods:** Primarily used with neural networks like Autoencoders, these methods learn to compress and then reconstruct normal data. Since they are trained on "normal" patterns, they poorly reconstruct anomalies, leading to a high reconstruction error which flags the outlier.

Q 6 : What is time series analysis? Mention two key components of time series
data.


Ans-

Time series analysis is a statistical method used to analyze data points collected sequentially over time. Its primary goal is to identify patterns, trends, and seasonal variations within the data. This analysis is crucial for forecasting future values based on the historical patterns observed. It is widely used in fields like economics, finance, and weather forecasting.

Two key components of time series data are:
1.  **Trend:** This is the long-term, overall direction of the data, showing a general upward, downward, or stable movement over an extended period.
2.  **Seasonality:** These are regular, predictable patterns that repeat over a fixed and known period, such as daily, monthly, or yearly cycles.

Q 7 Describe the difference between seasonality and cyclic behavior in time
series.

Ans-

Seasonality refers to predictable, fixed-period fluctuations that repeat every calendar year, driven by natural or human-made cycles. Examples include increased ice cream sales in summer or higher retail spending in December.

In contrast, cyclic behavior involves patterns that occur over periods longer than a year, and these periods are not fixed. These cycles are often tied to economic conditions, like booms and recessions, and their duration can vary significantly. While seasonality is strictly calendar-based and regular, cyclic patterns are longer-term and irregular in their timing.

Q 8 Write Python code to perform K-means clustering on a sample dataset.


```python
iAns-

import matplotlib.pyplot as plt
from sklearn.cluster import KMeans
from sklearn.datasets import make_blobs

# Generate sample dataset
X, y_true = make_blobs(n_samples=300, centers=4, cluster_std=0.60, random_state=0)

# Perform K-means clustering
kmeans = KMeans(n_clusters=4, random_state=0)
y_kmeans = kmeans.fit_predict(X)

# Plot the results
plt.scatter(X[:, 0], X[:, 1], c=y_kmeans, s=50, cmap='viridis')
centers = kmeans.cluster_centers_
plt.scatter(centers[:, 0], centers[:, 1], c='red', s=200, alpha=0.75, marker='X')
plt.title('K-means Clustering')
plt.show()
```

Q 9 What is inheritance in OOP? Provide a simple example in Python.


```python
Ans-

class Animal:
    def speak(self):
        return "Generic animal sound"

class Dog(Animal):  # Inherits from Animal
    def speak(self):  # Override parent method
        return "Woof!"

my_dog = Dog()
print(my_dog.speak())  # Output: Woof!
```

 Q 10 How can time series analysis be used for anomaly detection?

```python
Ans-

import numpy as np
import pandas as pd

# Generate sample time series data
dates = pd.date_range(start='2023-01-01', periods=100, freq='D')
values = np.random.normal(0, 1, 100)  # Normal data
values[30] = 8.5  # Add an anomaly

# Calculate rolling statistics
window = 20
df = pd.DataFrame({'value': values}, index=dates)
df['rolling_mean'] = df['value'].rolling(window=window).mean()
df['rolling_std'] = df['value'].rolling(window=window).std()

# Detect anomalies using Z-score (threshold=2)
df['zscore'] = (df['value'] - df['rolling_mean']) / df['rolling_std']
df['anomaly'] = np.abs(df['zscore']) > 2

print(df[df['anomaly']].index)
```
